```python
import numpy as np
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
def make_line(x, m, b, scatter):
    noise = scatter * np.random.randn(len(x))
    return x*m + b + noise, x*m + b, noise
```

```python
x = np.linspace(0, 1, 100)
y, y_noiseless, noise = make_line(x, 1, 2, .1)

plt.plot(x, y_noiseless, ls=':', lw=1, c='k')
plt.errorbar(x, y, noise, ecolor='k', fmt='o')
```
